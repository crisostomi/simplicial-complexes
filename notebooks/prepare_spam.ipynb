# Setup

## Imports

```python
import gzip
import os
import random

import dotenv
from pathlib import Path
from tqdm import tqdm
import networkx as nx
```

```python
random.seed(42)
```

## Paths

```python
dotenv.load_dotenv()

project_root = Path(os.environ["PROJECT_ROOT"])
data_path = project_root / "data" / "twitter-spam"
spammer_ids_path = data_path / "spam-ids.txt"
topology_path = data_path / "topology.gz"
subgraph_output_path = data_path / "output.txt"
```

# Collect subgraph

## Collect spammers and their neighbors

```python
spammer_ids = spammer_ids_path.read_text(encoding="utf-8").splitlines()
spammer_ids = set(spammer_ids)
```

```python
spammer_subsample = set(random.sample(spammer_ids, 100))
```

```python
subgraph_nodes = list(spammer_subsample)

with gzip.open(str(topology_path), mode="rt") as fr:
    for line in tqdm(fr):
        source, target = line.strip().split()
        
        if source in spammer_subsample or target in spammer_subsample:
            subgraph_nodes.append(source)
            subgraph_nodes.append(target)

subgraph_nodes = set(subgraph_nodes)

print(
    f"There are {len(subgraph_nodes)} nodes in the subgraph of spammers and their 1-neighborhood"
)

subgraph_nodes_str = "\n".join(subgraph_nodes)
```

```python
subgraph_output_path.write_text(data=subgraph_nodes_str, encoding="utf-8")
```

## Collect edges of the subsample 

```python
subgraph_nodes = subgraph_output_path.read_text().splitlines()
print(len(subgraph_nodes))
```

```python
subgraph_nodes = set(subgraph_nodes)
```

```python
G = nx.Graph()
G.add_nodes_from(subgraph_nodes)
```

```python
subgraph_edges = []
with gzip.open(str(topology_path), mode="rt") as fr:
    for line in tqdm(fr):
        source, target = line.strip().split()
        
        if source in subgraph_nodes and target in subgraph_nodes:
            edge = (source, target)
            subgraph_edges.append(edge)

print(f'There are {len(subgraph_edges)} edges in the induced subgraph')
```

```python
G.add_edges_from(subgraph_edges)
```
